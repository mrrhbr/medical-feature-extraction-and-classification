```python
import pandas as pd
import tensorflow as tf
from collections import Counter
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/Dataset/persica.csv'
df = pd.read_csv(file_path, on_bad_lines='skip')
```

```python
df = df.drop('Unnamed: 1', axis=1)
```

```python
df['News_text']=df['843656']
```

```python
df = df.drop('843656', axis=1)
```

```python
# Define a regular expression to remove digits and punctuation
df['News_text'] = df['News_text'].replace(to_replace=r'[^\w\s]', value=' ', regex=True)  # Remove punctuation
df['News_text'] = df['News_text'].replace(to_replace=r'\d+', value=' ', regex=True)  # Remove numbers
df['News_text'] = df['News_text'].replace(to_replace=r'[a-zA-Z0-9]', value=' ', regex=True)
```

```python
df = df[~df['News_text'].str.strip().eq('')]
```

```python
df = df.dropna(subset=['News_text'])
```

```python
all_text = ' '.join(df['News_text'])
```

```python
words = all_text.split()
```

```python
with tf.device('/GPU:0'):
    # Your code here
    word_dict = dict(Counter(words))
```

```python
dict_length = len(word_dict)
print("Length of word_dict:", dict_length)
```

```python
word_dict_str = str(word_dict)

# Save the string to a text file in Colab
with open('dict.txt', 'w') as file:
    file.write(word_dict_str)
```
